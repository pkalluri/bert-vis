## Load sample

```python
with open(tokens_path, 'rb') as f:
    tokens = pickle.load(f)

print('Tokens:')
print(tokens)

doc = ''.join(tokens)
print()
print('Document:')
print(doc)
```

## Visualize

```python
layer_to_reduced_acts = np.load(acts_path)
layer_to_colors = {layer:vis_util.channels_to_rgb(layer_to_reduced_acts[layer]) for layer in layer_to_reduced_acts.files}
layers = layer_to_reduced_acts.files
```

```python
layer_to_html = {}
for layer in layers:
    html = ''
    for tok, color in zip(tokens, layer_to_colors[layer]):
        rgba = "rgba(%s,%s,%s,1)" % (int(color[0]*255), int(color[1]*255), int(color[2]*255))
        html += f"<span style='background-color: {rgba};'>{tok}</span>"
    layer_to_html[layer] = html
```

```python
for layer in layers:
    print(layer)
    _display_html(layer_to_html[layer])
```
